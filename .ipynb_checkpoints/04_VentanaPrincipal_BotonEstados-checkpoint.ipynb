```python
'''
Vamos a comprobar los estados del boton con la 
funcion .setChecked
Para eso mantenemos las funciones anteriores del boton 
solo que 
cambiamos algunos nombres, ahora la nueva funcion es

def funcionEstadoActualBoton(estado):
    estadoBoton = estado


Le pondremos un solo parametro que sera el estado del boton

Este estado sera comprobado con la funcion setCheckable

boton.setChecked(estadoBoton)

Dentro del metodo implementado imprimimos el ultimo 
estado del boton

def funcionEstadoActualBoton(estado):
    estadoBoton = estado
    print(estadoBoton)
'''



import sys
from PyQt5.QtWidgets import QApplication, QPushButton

app = QApplication(sys.argv)

#=============== FUNCION BOTON ==================
#Con el parametro de la funcion capturamos el estado actual
estadoBoton=True

def funcionEstadoActualBoton(estado):
    estadoBoton = estado
    print("Estado Actual",estadoBoton)
 
#=============== BOTON ==================

#Instanciamos un objeto de la clase QPushButton
boton = QPushButton("Boton")

boton.resize(300,300)
boton.setWindowTitle('Ventana Boton')

#Habilitamos el cambio de estado
boton.setCheckable(True)
boton.clicked.connect(funcionEstadoActualBoton)

#=============== EJECUCION ==================
boton.show()

app.exec_()

```
